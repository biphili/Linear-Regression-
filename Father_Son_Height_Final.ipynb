```python
import pandas as pd 
```

```python
df=pd.read_csv('Father_Son_height_Corrected.csv')
```

```python
df
```

```python
import numpy as np 
```

```python
x_train=df['Father'].values[:,None]
x_train
x_train.shape
```

```python
y_train=df['Son'].values[:,None]
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
x_test=[[72.8],[61.1],[63.1],[68.1],[65]]
```

```python
predictions=lm.predict(x_test)
print(predictions)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(x_train,y_train,color='b')
plt.plot(x_test,predictions,color='black',linewidth=3)
plt.xlabel('Father height in inches')
plt.ylabel('Son height in inches')
plt.show()
```
